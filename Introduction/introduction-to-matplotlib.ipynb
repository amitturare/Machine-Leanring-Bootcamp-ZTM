### Two options for subplot


```python
# Subplot Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots Option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot or matplotlib OO method?)
* When plotting something quickly, okay to use the pyplot method.
* When plotting something more advance, use OO method.

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50.head()
```

```python
# Pyplot Method
over_50.plot.scatter(x='age', y='chol', c='target');
```

```python
# OO Method misxed with Pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot.scatter(x='age', y='chol', c='target', ax=ax)
# ax.set_xlim([45, 100]) # Adjust the limit on the x-axis
```

```python
over_50.target.value_counts()
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)

# Customize the plot
ax.set(title='Heart Disease and Cholestrol Levels', xlabel='Age', ylabel='Cholestrol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(y=over_50.chol.mean(), color='r', linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age and thalach
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax[0]
scatter = ax[0].scatter(x=over_50.age, y=over_50.chol, c=over_50.target)

# Customize the plot ax[0]
ax[0].set(title='Heart Disease and Cholestrol Levels', ylabel='Cholestrol')

# Add a legend
ax[0].legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax[0].axhline(y=over_50.chol.mean(), color='r', linestyle='--')


# Add data to ax[1]
ax[1].scatter(x=over_50.age, y=over_50.thalach, c=over_50.target)

# Customize the plot ax[1]
ax[1].set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')

# Add a legend
ax[1].legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax[1].axhline(over_50.thalach.mean(), c='r', ls='--')

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# Make the figure and the ax
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap='winter') # "cmap=''" Changes the color scheme

# Label the graph
ax.set(title='Heart Disease and Cholestrol Control', xlabel='Age', ylabel='Cholestrol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal meanline
ax.axhline(over_50.chol.mean(), c='r', ls='--');
```

```python
# Customize the y and x axis limitations


# Subplot of chol, age and thalach
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax[0]
scatter = ax[0].scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap='winter')

# Customize the plot ax[0]
ax[0].set(title='Heart Disease and Cholestrol Levels', ylabel='Cholestrol')

# Change the x axis limits
ax[0].set_xlim([50, 80])

# Add a legend
ax[0].legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax[0].axhline(y=over_50.chol.mean(), color='r', linestyle='--')


# Add data to ax[1]
ax[1].scatter(x=over_50.age, y=over_50.thalach, c=over_50.target, cmap='winter')

# Customize the plot ax[1]
ax[1].set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')

# Change the y axis limits
ax[1].set_ylim([60, 200])

# Add a legend
ax[1].legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax[1].axhline(over_50.thalach.mean(), c='r', ls='--')

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
